Under Sampling 

Building a sample dataset of similar distribution of legit transactions

no. of fraudulent trans --> 492 


```python
l_sample = l.sample(n = 492)
```

```python
#Concatenating 
new = pd.concat([l_sample,f],axis = 0)
```

```python
new.head()
```

```python
new.tail()
```

```python
new.hist(edgecolor='black',linewidth=1.2)
fig=plt.gcf()
fig.set_size_inches(23,20)
plt.show()
```

```python
new['Class'].value_counts()
```

```python
new.groupby('Class').mean()
```

```python
#Splitting data into features and target
x = new.drop(columns = 'Class', axis = 1)
y = new['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
#Splitting data into testing and training data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2,stratify = y,random_state =2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

 **MODEL** 

```python
model = LogisticRegression( max_iter = 990)
```

```python
model.fit(x_train,y_train)
```

***Model Evaluation***

```python
#Accuracy on train data
x_train_predict = model.predict(x_train)
```

```python
print(x_train_predict)
```

```python
train_data_accuracy = accuracy_score(y_train,x_train_predict)
print('Accuracy score of training data : ',train_data_accuracy)
```

```python
#Accuracy on test data
x_test_predict = model.predict(x_test)
```

```python
print(x_test_predict)
```

```python
#Accuracy on train data
test_data_accuracy = accuracy_score(y_test,x_test_predict)
print('Accuracy score of testing data : ',test_data_accuracy)
```
